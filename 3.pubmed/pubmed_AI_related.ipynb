```python
import os
import numpy as np
import pandas as pd
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.ssl_ import create_urllib3_context
from bs4 import BeautifulSoup
import time
```

```python
# solve SLLerror and requests error(caused by prox)
CIPHERS = (
    'ECDH+AESGCM:DH+AESGCM:ECDH+AES256:DH+AES256:ECDH+AES128:DH+AES:ECDH+HIGH:'
    'DH+HIGH:ECDH+3DES:DH+3DES:RSA+AESGCM:RSA+AES:RSA+HIGH:RSA+3DES:!aNULL:'
    '!eNULL:!MD5'
)

class DESAdapter(HTTPAdapter):
    """
    A TransportAdapter that re-enables 3DES support in Requests.
    """
    def init_poolmanager(self, *args, **kwargs):
        context = create_urllib3_context(ciphers=CIPHERS)
        kwargs['ssl_context'] = context
        return super(DESAdapter, self).init_poolmanager(*args, **kwargs)

    def proxy_manager_for(self, *args, **kwargs):
        context = create_urllib3_context(ciphers=CIPHERS)
        kwargs['ssl_context'] = context
        return super(DESAdapter, self).proxy_manager_for(*args, **kwargs)

s = requests.Session()
s.mount('https://api.rxivist.org', DESAdapter())
# headers = {'User-Agent': 'User-Agent:Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
# headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5) AppleWebKit 537.36 (KHTML, like Gecko) Chrome",
# "Accept": "text/html,application/xhtml+xml,application/xml; q=0.9,image/webp,*/*;q=0.8"}


headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.103 Safari/537.36'}
```

```python
queryKey = searchParse.find('QueryKey').text
webenv = searchParse.find('WebEnv').text
count = int(searchParse.find('Count').text)
retmax = int(searchParse.find('RetMax').text)
print("query: %s , webenv: %s , total counts: %d , retmax: %d"%(queryKey,webenv,count,retmax))
```

```python
searchList = ["PubDate" , "Source" , "LastAuthor" , "Title" , 
              "FullJournalName" , "ELocationID" ]
outDf = pd.DataFrame(columns=["idx","publish_date","source","last_author",
                              "title","full_jounal_name","elinks"])
```

```python
def grepPageRecords(retmax , webenv , currentNum):
    originDf = pd.DataFrame(columns=["idx","publish_date","source","last_author",
                              "title","full_jounal_name","elinks"])
    
    url2 = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esummary.fcgi?query_key=1&retmax="\
    +str(retmax)+"&WebEnv="+webenv+"&retstart="+str(currentNum)
    summaryRes = s.get(url2 , headers = headers)
    summaryParse = BeautifulSoup(summaryRes.text , 'xml')

    DocSumL = summaryParse.find_all("DocSum")

    for docSum in DocSumL:
        idx = docSum.find("Id" , recursive = False).string

        originDf = originDf.append(pd.DataFrame(
        {"idx" : idx , 
         "publish_date" : docSum.find("Item" , Name = searchList[0]).string ,
         "source" :  docSum.find("Item" , Name = searchList[1]).string , 
         "last_author" : docSum.find("Item" , Name = searchList[2]).string ,
         "title" : docSum.find("Item" , Name = searchList[3]).string,
         "full_jounal_name" : docSum.find("Item" , Name = searchList[4]).string,
         "elinks" : docSum.find("Item" , Name = searchList[5]).string
        }, index = [0]
        ), ignore_index=True)
    
    time.sleep(3)

    return(originDf)
```

```python
for retstart in range(0 , count , retmax):
    tmpDf = grepPageRecords(retmax , webenv , retstart)
    outDf = pd.concat([outDf , tmpDf])

print("%s, with search counts number is %d and output dataframe shape row is %d"%
      (["Number wrong!!!","Number right."][count == int(outDf.shape[0])],count,int(outDf.shape[0])))
outDf1 = outDf.reset_index(drop=True) 

outDf1.to_csv("./pubmed_AIrelated_papers_vFull_names.tsv" , sep="\t" , index=0)
```
